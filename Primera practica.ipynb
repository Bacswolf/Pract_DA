```python
#Librerias
import pandas as pd
import numpy as np
```

```python
#ruta del archivo
file = 'data/challenge/SN_d_tot_V2.0.txt'
```

```python
#leer el archivo
data = pd.read_csv(file, sep='\s+', header=None, usecols=range(0, 7))
data
# Primer error encontrado en numero de columnas suele ser 7 pero apartir de la fila 74966 se vuelve 8 (esto pasa porque en el archivo hay un *) por lo que al investigar encontre que 
# (error_bad_lines=False) skipea las filas que no tienen el mismo numero de columnas pero perdemlos esos datos
# otra opcio es usecols=range(0, n) es para seleccionar las columnas que se quieren leer por lo que nos sirve para figar el numero de columnas que necesitamos
```

Al leer el archivo podemos observar que todas las columna no tienen encabezado, por lo que se procede a agregarlos.
Recordemos que es una de las peticiones de la tarea.
- **txt file has no column headers**

```python
#definiendo encabezados
#Para definir los encabezados se puede obzerban en el link: https://www.sidc.be/SILSO/home
#se obzerva que esta por fecas y que los valores por dia tienen un maximo y un minimo que son las lineas de color amarillo
encabezado = ['año', 'mes', 'dia', 'dec_feca', 'max_sunspots','min_sunspots', 'definite']
encabezado
```

```python

data = pd.read_csv(file, sep='\s+' ,header=None, names=encabezado, usecols=range(0, 7))
data
```

```python
data.info()
#Podemos ver que no hay datos vacios
#año, mes y dia son enteros por lo que tenemos que juntarlos y convertirlos en fecha
```

```python
#explorando el archivo
data.head()
#existen valores negativos en los maximos y minimos de los sunspots por lo que se deben de eliminar
```

```python
data.tail()
```

```python
#eliminando los valores negativos
data = pd.read_csv(file, sep='\s+' ,header=None, names=encabezado, usecols=range(0, 7),na_values='-1')
data
#segundo error encontrado los sunspots maximos y minimos tienen valores negativos por lo que se deben de eliminar
#solucion aplicar na_values='-1' para que los valores negativos se vuelvan NaN
```

```python
data.info()
#los maximos y minimos de los sunspots son menos valores ya que converitomos los valores negativos en NaN
```

```python
data.describe()
#confirmamos que no hay valores negativos en los maximos y minimos de los sunspots
```

```python
data = pd.read_csv(file, sep='\s+' ,header=None, names=encabezado, usecols=range(0, 7), na_values='-1', parse_dates=[[0, 1, 2]])
data
#tercer error encontrado los valores de año, mes y dia no estan en formato fecha por lo que se deben de unir y convertir en fecha
#solucion aplicar parse_dates=[[0, 1, 2]] para que se unan y se conviertan en una sola columna de fecha
```

```python
data.info()
#Podemos ver que ahora la columna de año_mes_dia es de tipo fecha
```
